#### Graph Functions
1. plot                     -Simple Line Plot
2. plot.area                -Filled Version of line plot
3. plot.bar / plot.barh     -Bar Graph 
4. plot.box                 -Box Plot Density Line (Smooth Curved)
5. plot.density             -Density Line(Smooth Curved)
6. plot.hexbin              -Hexagon Version of Scatter 
7. plot.hist                -Histogram
8. plot.kde                 -Kernel Density Estimator(Density Graph)
9. plot.pie                 -Pie Chart of Categorical Data
10. plot.scatter            -Scatter Graph

```python
df2.plot()
df2.plot.area()
df2.plot.bar()
df2.plot.box()
df2.plot.density()
df2.plot.hexbin(x='a', y='b',C='c')
df2.plot.hist()
df2.plot.kde()
# df2.plot.pie()
df2.plot.scatter(x='a',y='b')
```

# Seaborn

```python
import seaborn as sns
```

###### Categories
1. Categorical Plots(hue can be used)
    a. countplot(single param)
    b. barplot(one categorical , one numerical and calculates mean by defualt)
    c. boxplot
    d. violinplot
    e. swarmplot
2. Distribution Plots
    a. distplot
    b. displot
    c. jointplot
    d. pairplot
    e. kdeplot
    f. rugplot
    g, Scatterplot
3. Matrix Plots
    a. heatmap
    b. clusterplot
    c. pairplot (same as above one)
4. Regression Plot
    a. implot
    b. regplot
5. Grid Plot
    a. PairGrid
    b. JointGrid
    c. FacetGrid

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.describe(include='all')
```

```python
tips.value_counts('day')
```

```python
tips.groupby(['sex','time'])['day'].value_counts().sort_index()
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.countplot(x='day',data=tips,hue='sex')
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=sum)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='time',estimator=np.mean)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='time',estimator=np.mean)
```
